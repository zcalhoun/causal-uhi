# Bootstrapped Results
This notebook is used to analyze the results of the bootstrapped script, and the results correspond to the table in the manuscript.

To use this notebook, you must have triggered a bootstrapped experiment.

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
# No iter
df = pd.read_csv('../results/bootstrap_final/results.csv')
df = df[df['val_score'] > 0]
print(df['val_score'].mean())
print(df['train_score'].mean())
```

```python
# Print out the coefficients for the no iteration
# Let's look at the model coefficien
plt.boxplot(df.values[:, 3:-3])
plt.xlabel("Bootstrapped Model Coefficients")
plt.ylabel("Coefficient Value")
# plt.grid(axis='y')
plt.axhline(y=0, color='r', linestyle='--', alpha=0.5)
plt.show()
```

```python
print("coef", "mean", "95% CI", sep='\t')
print("---------")
for i, row in enumerate(df.values[:, 3:-3].T):
    print(i, f"{row.mean():.2f}", f"({np.percentile(row, 2.5):.2f}, {np.percentile(row, 97.5):.2f})", sep='\t')
```
